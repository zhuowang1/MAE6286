```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
T=100.0
dt=0.02
N=int(T/dt)+1
```

```python
t=numpy.linspace(0.0,T,num=N)
```

```python
z0=100.0
b0=10.0
zt=100.0
g=9.81

u=numpy.array([z0,b0])

z=numpy.zeros(N)
z[0]=z0
```

```python
for n in range(1, N):
    f_u = numpy.array([u[1], g*(1-u[0]/zt)])
    u=u+dt*f_u
    z[n]=u[0]
```

```python
pyplot.figure(figsize=(9.0,4.0))
pyplot.xlabel('Time [s]')
pyplot.ylabel('Elevation [m]')
pyplot.xlim(t[0],t[-1])
pyplot.ylim(40.0, 160.0)
pyplot.grid()
pyplot.plot(t,z,color='C0',linestyle='-', linewidth=2)
```

```python
z_exact = (b0 * (zt / g)**0.5 * numpy.sin((g / zt)**0.5 * t) +
           (z0 - zt) * numpy.cos((g / zt)**0.5 * t) + zt)
```

```python
pyplot.figure(figsize=(9.0, 4.0))  
pyplot.title('Elevation of the phugoid over the time') 
pyplot.xlabel('Time [s]')  
pyplot.ylabel('Elevation [m]')  
pyplot.xlim(t[0], t[-1])  
pyplot.ylim(40.0, 160.0)  
pyplot.grid() 
pyplot.plot(t, z, label='Numerical',
            color='C0', linestyle='-', linewidth=2)
pyplot.plot(t, z_exact, label='Analytical',
            color='C1', linestyle='-', linewidth=2)
pyplot.legend(); 

```

```python
dt_values = [0.1, 0.05, 0.01, 0.005, 0.001, 0.0001]


z_values = []

for dt in dt_values:
    N = int(T / dt) + 1  
    t = numpy.linspace(0.0, T, num=N)  
    u = numpy.array([z0, b0])
    z = numpy.empty_like(t)
    z[0] = z0
    for n in range(1, N):
        rhs = numpy.array([u[1], g * (1 - u[0] / zt)])
        u = u + dt * rhs
        z[n] = u[0]  
    z_values.append(z)  
```

```python
def l1_error(z, z_exact, dt):
    """
    Computes and returns the error
    (between the numerical and exact solutions)
    in the L1 norm.
    
    Parameters
    ----------
    z : numpy.ndarray
        The numerical solution as an array of floats.
    z_exact : numpy.ndarray
        The analytical solution as an array of floats.
    dt : float
        The time-step size.
        
    Returns
    -------
    error: float
        L1-norm of the error with respect to the exact solution.
    """
    error = dt * numpy.sum(numpy.abs(z - z_exact))
    return error
```

```python
error_values = []

for z, dt in zip(z_values, dt_values):
    N = int(T / dt) + 1  
    t = numpy.linspace(0.0, T, num=N) 
    z_exact = (b0 * (zt / g)**0.5 * numpy.sin((g / zt)**0.5 * t) +
               (z0 - zt) * numpy.cos((g / zt)**0.5 * t) + zt)
    error_values.append(l1_error(z, z_exact, dt))
```

```python
pyplot.figure(figsize=(6.0, 6.0))
pyplot.title('L1-norm error vs. time-step size') 
pyplot.xlabel('$\Delta t$')  
pyplot.ylabel('Error')  
pyplot.grid()
pyplot.loglog(dt_values, error_values,
              color='C0', linestyle='--', marker='o')  
pyplot.axis('equal'); 
```

```python
def euler(u0,dt,N):
    u=numpy.zeros([N,2])
    u[0]=u0
    for n in range(1, N):
        rhs = numpy.array([u[n,1], g * (1 - u[n,0] / zt)])
        u[n] = u[n-1] + dt * rhs
    return u

dt_values = [0.1, 0.05, 0.01, 0.005, 0.001, 0.0001]


z_values = []

for dt in dt_values:
    N = int(T / dt) + 1  
    u0 = numpy.array([z0, b0])
    u=euler(u0,dt,N)
    z=u[:,0]
    z_values.append(z)  
```

```python
error_values = []

for z, dt in zip(z_values, dt_values):
    N = int(T / dt) + 1  
    t = numpy.linspace(0.0, T, num=N) 
    z_exact = (b0 * (zt / g)**0.5 * numpy.sin((g / zt)**0.5 * t) +
               (z0 - zt) * numpy.cos((g / zt)**0.5 * t) + zt)
    error_values.append(l1_error(z, z_exact, dt))
```

```python
pyplot.figure(figsize=(6.0, 6.0))
pyplot.title('L1-norm error vs. time-step size') 
pyplot.xlabel('$\Delta t$')  
pyplot.ylabel('Error')  
pyplot.grid()
pyplot.loglog(dt_values, error_values,
              color='C0', linestyle='--', marker='o')  
pyplot.axis('equal'); 
```
